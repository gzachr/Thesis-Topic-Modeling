```python
import spacy
from gliner_spacy.pipeline import GlinerSpacy
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("gliner_spacy", config={"labels": ["person", "location", "food"]})
```

```python
with open("HOSBltLGm0Q_captions.txt", "r", encoding="utf-8") as f:
    text = f.read()
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```
