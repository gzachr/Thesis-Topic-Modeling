```python
import os
import pandas as pd
import shutil
import random
import re

# Define paths
csv_path = "cleaned_results.xlsx"
transcripts_folder = "cleaned_transcripts"
destination_folder = "standard_dataset"

# Read the Excel file
data = pd.read_excel(csv_path)

# Filter rows where "related" column has the value "yes"
filtered_data = data[data["related"] == "yes"]

# Get the list of video IDs and corresponding titles
video_info = filtered_data[["Video Id", "Video Title"]].dropna().to_dict(orient="records")

# Ensure destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Define the number of samples you want (adjustable)
num_samples = 100  

# Function to sanitize filenames
def sanitize_filename(name, max_length=100):
    name = re.sub(r'[\/:*?"<>|]', "", name)  # Remove invalid characters
    return name[:max_length]  # Truncate if too long

# Get existing files in the destination folder (to prevent duplicates)
existing_files = set(os.listdir(destination_folder))

# Shuffle and iterate through video info to select valid samples
valid_samples = []
random.shuffle(video_info)

for entry in video_info:
    video_id = entry["Video Id"]
    video_title = entry["Video Title"]
    sanitized_title = sanitize_filename(video_title)
    destination_file = f"{video_id}_{sanitized_title}_captions.txt"

    source_file = os.path.join(transcripts_folder, f"{video_id}_captions.txt")
    destination_path = os.path.join(destination_folder, destination_file)
    
    # Skip if file already exists
    if destination_file in existing_files:
        continue  

    if os.path.exists(source_file):
        # Read the file and check word count
        with open(source_file, "r", encoding="utf-8") as file:
            content = file.read().strip()
            word_count = len(content.split())
        
        if word_count >= 100:
            valid_samples.append((video_id, sanitized_title, source_file, destination_path))
        
        if len(valid_samples) >= num_samples:
            break  # Stop once we reach the required sample size

# Copy the selected files with sanitized video title in the filename
for video_id, sanitized_title, source_file, destination_path in valid_samples:
    shutil.copy(source_file, destination_path)
    print(f"Copied: {source_file} -> {destination_path}")

print(f"\nSampling and copying completed! {len(valid_samples)} new files copied.")
```
