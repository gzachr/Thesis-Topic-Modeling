```python
# Store topics as list of lists
import json
# Load the JSON file
filename = 'T3.nodes.json'
filepath = '../output-jsons/' + filename
with open(filepath, 'r') as f:
    data = json.load(f)

def flatten_tree(node, depth=1, result=None):
    if result is None:
        result = []
    """Recursively flattens a tree structure into a list."""
    result.append({
        "id": node["id"],
        "texts": node["text"].split(),
        "level": depth
    })
    
    for child in node.get("children", []):
        flatten_tree(child, depth + 1, result)
    
    return result

# Process each root node in the JSON
flat_list = []
for root in data:
    flat_list.extend(flatten_tree(root))

print(flat_list)
```

```python
import pandas as pd

# Create the dataframe from flat_list
df = pd.DataFrame(flat_list)

# Rename columns as needed
df = df.rename(columns={'id': 'Id', 'level': 'level', 'texts': 'texts'})

# Add empty columns for 'general category' and 'specific category'
df['general category'] = ''
df['specific category'] = ''

df = df[['Id', 'level', 'texts', 'general category', 'specific category']]

df['texts'] = df['texts'].apply(lambda x: ' '.join(x))

df.columns = [col.title() for col in df.columns]
df.head()

df.to_csv('categories.csv', index=False)
```
